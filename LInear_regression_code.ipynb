# Linear Regression

Data: Kaggle
5000 rows * 7 columns

Here “price” is the dependent variable and 'Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population' are the independent variables. 

The variables are: 

• 'Avg. Area Income': Avg. Income of householder of the city house is located in. 

• 'Avg. Area House Age': Avg. Age of Houses in same city. 

• 'Avg. Area Number of Rooms': Avg. Number of Rooms for Houses in same city. 

• 'Avg. Area Number of Bedrooms': Avg. Number of Bedrooms for Houses in same city. 

• 'Area Population': Population of city. 

• 'Price': Price that the house sold at. 

• 'Address': Address of the houses. 

```python
import pandas as pd
import numpy as np
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
house = pd.read_csv('/home/amma/Downloads/USA_Housing.csv')
house
```

```python
house.info()
```

```python
house.columns
```

```python
x = house[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]

y = house['Price']
```

# Distribution of target variable 

```python
plt.figure(figsize=(16,6))
sns.distplot(house.Price)
plt.show()
```

# Model training

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 121)
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

y intercept

```python
print(lm.intercept_)
```

```python
y_pred = lm.predict(x_test)
```

```python
plt.scatter(y_test, y_pred)
```

```python
mse = metrics.mean_squared_error(y_test, y_pred)
mae = metrics.mean_absolute_error(y_test, y_pred)
r2= metrics.r2_score(y_test, y_pred)
```

```python
print('MSE: ', mse)
print('MAE: ', mae)
print('R2: ', r2)
```

Conclusions: 

In the above scatter plot, we see data is in line shape, which means the model has done good predictions. 

The model achieved a Mean Squared Error (MSE) of 10,160,806,230.59, which is reasonable considering the large scale of house prices. The Mean Absolute Error (MAE) of 80,804.88 indicates that, on average, the predicted house prices differ from the actual prices by approximately 80,805 units, reflecting good prediction accuracy. 

The R² score of 0.9226805 shows that 92.26% of the variation in house prices is explained by the independent variables. This high R² value confirms a strong linear relationship between the predictors and the target variable.
